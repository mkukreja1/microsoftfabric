```python
from delta.tables import *
import datetime 
import time
```

```python
table_name = ""
database = ""
incremental_column = ""
delta_source_path = ""
```

```python
base_path = f"abfss://Training@onelake.dfs.fabric.microsoft.com/trg_lakehouse.Lakehouse/Files"
path = f"{base_path}/{delta_source_path}"
database = "trg_lakehouse"
bronze_table_name = f"bronze_{table_name}"

display(base_path)
display(delta_source_path)
display(path)
display(database)
display(table_name)
display(bronze_table_name)
display(incremental_column)
```

```python
df = spark.read.csv(path, header="True", inferSchema="True")

if df.count() > 0:
    # df.write.format("delta").mode("append").saveAsTable(bronze_table_name)
    df.write.format("delta").mode("append").option("mergeSchema","True").saveAsTable(bronze_table_name)
```

```python
query = f"""UPDATE bronze_control_table 
               SET incremental_column_value = (SELECT to_timestamp(max({incremental_column})) FROM {database}.{bronze_table_name})
             WHERE table_name = '{table_name}'"""
spark.sql(query)
```
