```python
df = spark.sql("SELECT * FROM Electroniz_Lakehouse.bronze_store_customers")
display(df)
```

```python
# Code generated by Data Wrangler for PySpark DataFrame

from pyspark.sql import functions as F

def clean_data(df):
    # Replace all instances of "+" with "" in column: 'phone'
    df = df.withColumn('phone', F.regexp_replace('phone', "(?i)\\+", ""))
    # Replace all instances of " " with "" in column: 'phone'
    df = df.withColumn('phone', F.regexp_replace('phone', "(?i) ", ""))
    # Replace all instances of "(" with "" in column: 'phone'
    df = df.withColumn('phone', F.regexp_replace('phone', "(?i)\\(", ""))
    # Replace all instances of ")" with "" in column: 'phone'
    df = df.withColumn('phone', F.regexp_replace('phone', "(?i)\\)", ""))
    # Replace all instances of "-" with "" in column: 'phone'
    df = df.withColumn('phone', F.regexp_replace('phone', "(?i)-", ""))
    # Convert text to lowercase in column: 'email'
    df = df.withColumn('email', F.lower(F.col('email')))
    return df

df_clean = clean_data(df)
display(df_clean)
```

```python
df_clean.write.format("delta").saveAsTable("bronze_clean_store_customers")
```
