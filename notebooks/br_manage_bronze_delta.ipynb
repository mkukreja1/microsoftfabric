```python
from delta.tables import *
import datetime 
import time
```

```python
table_name = ""
domain_name = ""
incremental_column = ""
delta_lakehouse_path = ""
lakehouse = ""
```

```python
path = f"Files/{delta_lakehouse_path}"
bronze_table_name = f"bronze_{table_name}"


display(delta_lakehouse_path)
display(path)
display(domain_name)
display(table_name)
display(bronze_table_name)
display(incremental_column)
display(lakehouse)
```

```python
df = spark.read.csv(path, header="True", inferSchema="True")

if df.count() > 0:
    df.write.format("delta").mode("append").option("mergeSchema","True").saveAsTable(bronze_table_name)
```

```python
query = f"""UPDATE bronze_watermark_table 
               SET incremental_column_value = (SELECT to_timestamp(max({incremental_column})) FROM {lakehouse}.{bronze_table_name})
             WHERE table_name = '{table_name}'"""
spark.sql(query)
```
